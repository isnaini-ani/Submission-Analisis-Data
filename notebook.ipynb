### Cleaning Data

```python
#mengganti tipe beberapa kolom menjadi datetime di data order 
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date",
"order_delivered_customer_date", "order_estimated_delivery_date"]
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana demografi pelanggan yang ada?

```python
customers_df.info()
```

```python
#menampilkan demografi customer berdasarkan kota
bycity_df = all_df.groupby(by="customer_city").customer_id.nunique().reset_index()
bycity_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bycity_df = bycity_df.sort_values(by="customer_count", ascending=False).head(10)
bycity_df
plt.figure(figsize=(10, 5))
colors_ = ["#72BCD4"] + ["#D3D3D3"] * (len(bycity_df) - 1)
sns.barplot(
    x="customer_count",
    y="customer_city",
    data=bycity_df.sort_values(by="customer_count", ascending=False),
    hue="customer_city",
    palette=colors_,
    legend=False 
)
plt.title("Jumlah Pelanggan berdasarkan Kota", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

```python
#menampilkan demografi customer berdasarkan state
bystate_df = all_df.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bystate_df = bystate_df.sort_values(by="customer_count", ascending=False).head(10)
bystate_df
plt.figure(figsize=(10, 5))
colors_ = ["#72BCD4"] + ["#D3D3D3"] * (len(bystate_df) - 1)
sns.barplot(
    x="customer_count",
    y="customer_state",
    data=bystate_df.sort_values(by="customer_count", ascending=False),
    hue="customer_state",
    palette=colors_,
    legend=False 
)
plt.title("Jumlah Pelanggan berdasarkan State", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

### Pertanyaan 2: Bagaimana performa penjualan dalam beberapa bulan terakhir?

```python
#banyaknya customer yang melakukan transaksi
cus_2018_df = all_df[all_df["order_purchase_timestamp"].dt.year == 2018]
customer_monthly_df = cus_2018_df.resample(rule='ME', on='order_purchase_timestamp').agg({
    "customer_id": "nunique"
})
customer_monthly_df.index = customer_monthly_df.index.strftime('%B')

customer_monthly_df = customer_monthly_df.reset_index()
customer_monthly_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

customer_monthly_df.head()

plt.figure(figsize=(10, 5)) 
plt.plot(customer_monthly_df["order_purchase_timestamp"], customer_monthly_df["customer_count"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Banyaknya customer yang melakukan transaksi per-Bulan (2018)", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

```python
#banyaknya order yang dilakukan customer
order_2018_df = all_df[all_df["order_delivered_customer_date"].dt.year == 2018]
datetime_order_monthly_df = order_2018_df.resample(rule='ME', on='order_delivered_customer_date').agg({
    "order_id": "nunique"
})
datetime_order_monthly_df.index = datetime_order_monthly_df.index.strftime('%B')

datetime_order_monthly_df = datetime_order_monthly_df.reset_index()
datetime_order_monthly_df.rename(columns={
    "order_id": "order_count"
}, inplace=True)

datetime_order_monthly_df.head()

plt.figure(figsize=(10, 5)) 
plt.plot(datetime_order_monthly_df["order_delivered_customer_date"], datetime_order_monthly_df["order_count"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Jumlah Order per-Bulan (2018)", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

**Insight:**
- jumlah order yang paling banyak pada bulan Agustus
- jumlah customer paling banyak yang melakukan transaksi berada pada maret
